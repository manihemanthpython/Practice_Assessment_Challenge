Task 2: Apply one-hot encoding to any categorical column in the dataset, split the data into training and test sets (80/20), and then scale the numerical features using StandardScaler — ensuring scaling is applied after the split.

```python
# Apply one-hot encoding to any categorical column in the dataset
print(df['seller_type'].value_counts())
print(df['seller_type'].unique())
```

```python
print(df['fuel_type'].value_counts())
print(df['fuel_type'].unique())
```

```python
print(df['transmission_type'].value_counts())
print("\n",df['transmission_type'].unique())
```

```python
print(df['brand'].value_counts())
print(df['brand'].unique())
```

```python
print(f"Before Encoding: {df.shape}")
df = pd.get_dummies(df,columns=['transmission_type'])
print(f"After Encoding: {df.shape}")
```

```python
print(f"Before Encoding: {df.shape}")
df = pd.get_dummies(df, columns=['seller_type', 'fuel_type', 'brand'], drop_first=True)
print(f"After Encoding: {df.shape}")
```

```python
# split the data into training and test sets (80/20)
x = df.drop(columns='selling_price')
y = df['selling_price']

x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.80, random_state= 42)

print(f"Train: {x_train.shape[0]} rows ({x_train.shape[0]/len(x)*100:.0f}%)")
print(f"Test:  {x_test.shape[0]} rows ({x_test.shape[0]/len(x)*100:.0f}%)")
```

```python
scaler = StandardScaler()
numeric_cols = ['vehicle_age', 'km_driven', 'mileage', 'engine', 'max_power', 'seats']
x_train[numeric_cols] = scaler.fit_transform(x_train[numeric_cols])
x_train
```
