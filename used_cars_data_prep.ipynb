For columns mileage, engine, and max_power — strip any unit strings (e.g. "kmpl", "CC", "bhp"), convert to numeric using pd.to_numeric(..., errors='coerce'), and fill remaining nulls with each column's median.

```python
# Drop rows where selling_price is missing.

df = df.dropna(subset=['selling_price'])
df['selling_price'].isna().sum() *100 /len(df) 
```

```python
df['engine'].unique()
```

```python
df.info()
```

```python
# For columns mileage, engine, and max_power — strip any unit strings (e.g. "kmpl", "CC", "bhp"), 
# convert to numeric using pd.to_numeric(..., errors='coerce'), and fill remaining nulls with each column's median.

df['mileage'].astype(str).str.strip().str.replace("kmpl","").str.replace("km","")
df['engine'].astype(str).str.strip().str.replace("CC","")
df['max_power'].astype(str).str.strip().str.replace("bhp","")

df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')
df['engine'] = pd.to_numeric(df['engine'], errors='coerce')
df['max_power'] = pd.to_numeric(df['max_power'], errors='coerce')

columns = ["mileage", "engine", "max_power"]
for col in columns:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
    
    print(f"Filling '{col}' missing values with median: {median_value}")


```

```python
df.info()
```

```python
print(f"Before cleaning: {df['km_driven'].unique()}\n")
df['km_driven'] = df['km_driven'].str.strip().str.replace('kms', '').str.replace('km', '').str.replace(',','').astype(int)
print(f"After cleaning: {df['km_driven'].unique()}")
```

```python
# Remove rows where selling_price is 999999999 or below 10000.
before = len(df)
df = df[df['selling_price'] != 999999999]
df = df[df['selling_price'] >=10000]
after = len(df)
print(f"Removed rows: {before-after}")
print(f"Remaing rows: {after}")
```

```python
# Drop all duplicate rows.
print("Before Drop DUplicates: ",df.duplicated().sum())
df = df.drop_duplicates()
print("Before Drop DUplicates:",df.duplicated().sum())
```

```python
# Apply label encoding to transmission_type — map Manual → 0 and Automatic → 1.
df['transmission_type'].value_counts()  # counting the transmission types
```

# Apply one-hot encoding to fuel_type and seller_type using pd.get_dummies() with drop_first=True.

```python
df['fuel_type'].unique()
```

```python
# Removing spaces and chares in (fuel type) label
df['fuel_type'] = df['fuel_type'].astype(str).str.strip().str.title()
df['fuel_type'].unique()
```

```python
# Removing Spaces,chars and Spellings. 
df['seller_type'].unique()
```

```python
spelling_corrections = {
    'Individuall' : 'Individual',
    'Individul' : 'Individual',
    'Trustmark Dealer': 'Trustmark Dealer',
    'Delaer': 'Dealer'
}
df['seller_type'] = df['seller_type'].astype(str).str.title().replace(spelling_corrections)
```

```python
df['seller_type'].unique()
```

```python
df['brand'].unique()
```

```python
df['brand'] = df['brand'].astype(str).str.strip().str.title()
```

```python
seats_mod = df['seats'].mode()[0]
df['seats'] = df['seats'].fillna(seats_mod)
```

```python
df.shape
```

```python
# Now Apply one-hot encoding to fuel_type and seller_type using pd.get_dummies() with drop_first=True.
df = pd.get_dummies(df, columns=['fuel_type', 'seller_type', 'brand'], drop_first=True)
df.head()
```

Why is drop_first=True used in one-hot encoding, and what does a row of all zeros in the new dummy columns represent?

Ans:- The reason for dropping the first column in to prevent perfect multicollinearity. The multicollinearity occurs when one indepented variable in a regression model can be linarly predicted from high degree of accuracy.
when we using (drop first = true) the first category encountered is removed and become the reference category.

Task 3 — Split and Compute Baseline MAE

Define X (all columns except selling_price) and y (selling_price). Drop any remaining non-numeric columns from X.

Split into 80% train and 20% test using train_test_split with random_state=42.

Compute the baseline MAE: predict the mean of y_train for every row in the test set and calculate the MAE against y_test.

Print: "Baseline MAE: ₹X" rounded to the nearest rupee.

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
# Drop any remaining non-numeric columns from X.
print(df.columns)
```

```python
df = df.drop(columns=['car_name', 'model'])
df.head()
```

```python
df.isna().sum()
```

```python
# Define X (all columns except selling_price) and y (selling_price). 

df.columns
x = df.drop(columns=['selling_price'])
y = df['selling_price']
```

```python
# Split into 80% train and 20% test using train_test_split with random_state=42.
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.80, test_size=0.20, random_state=42)
```

```python
# Compute the baseline MAE: predict the mean of y_train for every row in the test set and calculate the MAE against y_test.
model = LinearRegression()
model.fit(x_train,y_train)
y_train_pred = model.predict(x_train)
train_MAE = mean_absolute_error(y_train, y_train_pred)
print(f"Y train Predict value:{y_train_pred}\n")
print(f"Mean Absoluate Error value: {train_MAE}\n")

Avg_seals = df['selling_price'].mean()
df['base_pred'] = Avg_seals
base_MAE = mean_absolute_error(df['selling_price'], df['base_pred'])
print(f"Baseline MAE: ₹:- {base_MAE:.2f} ")
```
