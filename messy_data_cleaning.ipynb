Task 2 — Clean the Data Fix the issues identified: drop null target rows, impute missing input features, strip and lowercase the brand column, extract numeric values from the mileage column, and remove duplicates.

```python
# 1. Drop null target rows (assuming 'price' is the target)
df = df.dropna(subset=['selling_price'])
```

```python
df.info()
```

```python
# imputing missing input features.
columns = ['engine','max_power']
for col in columns:
    median_valu = df[col].median()
    df[col].fillna(median_valu)
```

```python
#  strip and lowercase the brand column.
df['brand'] = df['brand'].str.strip().str.lower()
df['brand'].unique()
```

```python
# extract numeric values from the mileage column, and remove duplicates.
df['mileage'] = df['mileage'].str.strip().str.replace("kmpl","")
df['mileage'] = pd.to_numeric(df['mileage'])
```

```python
# Remove duplicates
df = df.drop_duplicates().reset_index(drop=True)
df.duplicated().sum()
```

Task 3 — Compute Baseline MAE Build a baseline model that predicts the mean selling_price for every record. Calculate and print the MAE of this baseline on the full cleaned dataset.

```python
from sklearn.metrics import mean_absolute_error
mean_price = df['selling_price'].mean()
baseline_predicts = [mean_price]*len(df)
baseline_MAE = mean_absolute_error(df['selling_price'],baseline_predicts)
print(f"Mean price: {mean_price:.2f}")
print(f"Baseline predicts : {baseline_MAE:.2f}")
```
